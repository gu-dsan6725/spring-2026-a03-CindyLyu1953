# Part 1: Code Q&A System with Bash Tools

RAG system that answers questions about the mcp-gateway-registry codebase using bash tools (grep, find, cat, tree) for context retrieval.

```python
import sys
from pathlib import Path

# Add project root to path (works whether cwd is project root or notebooks/)
PROJECT_ROOT = Path.cwd() if (Path.cwd() / "mcp-gateway-registry").exists() else Path.cwd().parent
sys.path.insert(0, str(PROJECT_ROOT))

from src.part1_bash_tools import answer_question, classify_question

REPO_PATH = PROJECT_ROOT / "mcp-gateway-registry"
assert REPO_PATH.exists(), f"mcp-gateway-registry not found at {REPO_PATH}"
```

```python
TEST_QUESTIONS = [
    (1, "What Python dependencies does this project use?", "Simple"),
    (2, "What is the main entry point file for the registry service?", "Simple"),
    (3, "What programming languages and file types are used in this repository? (e.g., Python, TypeScript, YAML, JSON, Dockerfile, etc.)", "Simple"),
    (4, "How does the authentication flow work, from token validation to user authorization?", "Difficult"),
    (5, "What are all the API endpoints available in the registry service and what scopes do they require?", "Difficult"),
    (6, "How would you add support for a new OAuth provider (e.g., Okta) to the authentication system? What files would need to be modified and what interfaces must be implemented?", "Very Hard"),
]
```

```python
import time
results = []
for num, question, difficulty in TEST_QUESTIONS:
    print(f"\n{'='*60}")
    print(f"Q{num} ({difficulty}): {question}")
    print("="*60)
    qtype = classify_question(question)
    print(f"Classified as: {qtype.value}")
    answer = answer_question(question, REPO_PATH)
    print(f"\nAnswer:\n{answer}")
    time.sleep(12)  # Avoid Groq rate limit (TPM) between questions
    results.append((num, question, difficulty, answer))
```

```python
output_path = PROJECT_ROOT / "part1_results.txt"
with open(output_path, "w") as f:
    for num, question, difficulty, answer in results:
        f.write(f"Question {num} ({difficulty}): {question}\n")
        f.write(f"Answer:\n{answer}\n")
        f.write("\n" + "-"*60 + "\n")
print(f"Results saved to {output_path}")
```
